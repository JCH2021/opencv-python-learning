```python
import numpy as np
img = np.zeros((512,512,3),np.uint8)
line = cv2.line(img,(384,0),(510,128),(0,0,255),10)
#
#
plt.imshow(line)
```

```python
rectangle = cv2.rectangle(img,(100,200),(400,300),(255,255,0),5)
plt.imshow(rectangle)
```

```python
circle = cv2.circle(img,(200,400),63,(0,255,255),-1)
#
plt.imshow(circle)
```

```python
ellipse = cv2.ellipse(img,(150,100),(200,100),0,0,360,(127,255,144),-1)
#
#
plt.imshow(ellipse)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(img,'Drawing',(300,400),font,3,(254,127,233),2)
#
#
#
plt.imshow(text)
```
